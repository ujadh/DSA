## Assignment 5.2 (1 p)

Implement new methods to `LinkedList` class.

This Python code defines two classes: `Node` and `LinkedList`, implementing a basic **linked list** structure.
* `Node` represents a node in the linked list and
* `LinkedList` represents the linked list itself.

In this assignment `LinkedList` implementation allows creating a linked list, add nodes, display the contents, and performing an iterative search for a specific value.

Do the following task:
* The `LinkedList` class will store the following integers: `3, 31, 21, 52, 16, 91, 10, and 5`. Insert these integers into the linked list.
* Implement a search method called `iterative_search()` in the `LinkedList` class that takes the *Integer* to be searched for as an argument and reports whether this integer is found in the `LinkedList`.
* Implement `search_items_from_list()`calling `iterative_search()` method from LinkedList class.
* In test, you must add all items from `added_values` list to `linked_list`.

```python
# Class for linkedlist node
class Node:
    def __init__(self, data=None, next_item=None):
        self.data = data
        self.next = next_item


# Class for linkedlist
class LinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, data):
        new_node = Node(data)
        if self.head:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        else:
            self.head = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data)
            current = current.next

    def iterative_search(self, n):
        # Implementing iterative search to find given Integer value from the linked list
        current = self.head
        while current:
            if current.data == n:
                return True
            current = current.next
        return False


def search_items_from_list(to_find, llist):
    # Implementing search_items_from_list by calling iterative_search method from LinkedList class
    found_items = []
    for item in to_find:
        found_items.append(llist.iterative_search(item))
    return found_items


# Test
correct_answer_05_02 = [True, True, False]
added_values = [13, 31, 21, 52, 16, 91, 10, 5]
linked_list = LinkedList()

# Adding all items from added_values to LinkedList
for value in added_values:
    linked_list.add_node(value)

linked_list.display()

search_items = [5, 31, 22]
found_items = search_items_from_list(search_items, linked_list)
print(found_items)

# Assertion
assert found_items == correct_answer_05_02
print(f"answer {found_items} was OK!")
```

## Assignment 5.3 (1 p)

Sort linked list elements.

Create a function called `ask_linked_list_from_user()` that prompts the user to input the number of nodes in the linked list.
Subsequently, ask the user to enter the same number of strings to be saved into the linked list.

Perform also the following operations:
* Sort the order of the linked list using the bubble sort method. You must implement bubble sort in this assignment in `bubble_sort_for_linked_list()` function
* Create a method that returns the content of the linked list as a `list`.

```python
import unittest

class Node:
    def __init__(self, data=None, next_item=None):
        self.data = data
        self.next = next_item


class LinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, data):
        new_node = Node(data)
        if self.head:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        else:
            self.head = new_node

    def display(self):
        current = self.head
        while current is not None:
            if current.data is not None:
                print(current.data)
            current = current.next

    def to_list(self):
        new_list = []
        current = self.head
        while current is not None:
            if current.data is not None:
                new_list.append(current.data)
            current = current.next
        return new_list

    def iterative_search(self, n):
        current = self.head
        while current:
            if current.data == n:
                return True
            current = current.next
        return False

def bubble_sort_for_linked_list(a):
    # Implementing bubble sort for elements in the linked list
    if a.head is None:
        return
    else:
        outer_node = a.head
        while outer_node:
            inner_node = a.head
            while inner_node.next:
                if inner_node.data > inner_node.next.data:
                    inner_node.data, inner_node.next.data = inner_node.next.data, inner_node.data
                inner_node = inner_node.next
            outer_node = outer_node.next

def ask_linked_list_from_user():
    # Asking the user for the number of nodes and then entering the same number of strings
    linked_list = LinkedList()
    num_nodes = int(input("Enter the number of nodes for the linked list: "))
    for i in range(num_nodes):
        data = input(f"Enter data for node {i + 1}: ")
        linked_list.add_node(data)
    return linked_list

# Test
linked_list = ask_linked_list_from_user()
sorted_orig_list = sorted(linked_list.to_list())

bubble_sort_for_linked_list(linked_list)
sorted_llist = linked_list.to_list()

# Assertion
unittest.TestCase().assertListEqual(sorted_llist, sorted_orig_list)
print(f"answer {sorted_llist} was OK!")
```

## Assignment 5.4

Create **Double Linked List** that stores `Person` objects.

In a Doubly linked list, each node contains a data element and two pointers, one pointing to the next node in the sequence and another pointing to the previous node.
It allows for traversal in both directions, forward and backward, to perform operations such as insertion and deletion at both ends of the list.

You should create the following methods in the `DoubleLinkedList` class:
* `add_node()`: Add a new `Node` to Double linked list.
* `Insert_at_the_beginning()`: Insert a new `Node` at the beginning of Double linked list.
* `remove_node()`: Optional part is to remove a `Node` from the Double linked list. It's optional because the test program does not call this method.

Note! The test program has already been implemented, so please don't edit it.

```python
# Class for linkedlist node
class Node:
    def __init__(self, data=None):
        self.prev = None
        self.data = data
        self.next = None


# Class for linkedlist
class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def insert_at_the_beginning(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def display(self):
        current = self.head
        nodes = []
        print('\n')
        while current:
            nodes.append(str(current.data))
            current = current.next
        print(' <-> '.join(nodes) + '\n')

    def to_string(self):
        current = self.head
        nodes = []
        while current:
            nodes.append(str(current.data))
            current = current.next
        return ' : '.join(nodes)


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def name(self):
        return self.name

    def age(self):
        return self.age

    def __str__(self):
        return f"{self.name}, {self.age}"


# Test
llist = DoubleLinkedList()

llist.add_node(Person("Bob", 45))
llist.add_node(Person("Jill", 37))
llist.add_node(Person("Jane", 29))
llist.insert_at_the_beginning(Person("Jack", 59))

llist.display()

result = "Jack, 59 : Bob, 45 : Jill, 37 : Jane, 29"

assert llist.to_string() == result

print(f"answer '{llist.to_string()}' was OK!")
```
